##### Task 4 equation
\begin{cases}
        \begin{aligned}
            10.2 x_1 - 3.2 x_2 &= 149 \\
            -5.8 x_1 + 16 x_2 &= 83 \\
            10.3 x_1 + 7.3 x_2 &= 234 \\
        \end{aligned}
    \end{cases}

```python
x1 = np.linspace(-5, 5, 100)

x2 = (149 - 10.2 * x1) / -3.2

plt.plot(x1, x2, "c")

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("10.2 x1 – 3.2 x2 = 149")
plt.grid()
plt.show()
```

```python
x1 = np.linspace(-5, 5, 100)

x2 = (83 + 5.8 * x1) / 16

plt.plot(x1, x2, "y--")

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("–5.8 x1 + 16 x2 = 83")
plt.grid()
plt.show()
```

```python
x1 = np.linspace(-5, 5, 100)

x2 = (234 - 10.3 * x1) / 7.3

plt.plot(x1, x2, "m-.")

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("10.3 x1 + 7.3 x2 = 234")
plt.grid()
plt.show()
```

```python
x = np.linspace(-5, 5, 100)


equations = [
    {'equation': lambda x: (149 - 10.2 * x) / -3.2, 'label': '10.2 x1 – 3.2 x2 = 149', 'color': 'red', 'linestyle': 'dashed'},
    {'equation': lambda x: (83 + 5.8 * x) / 16, 'label': '– 5.8 x1 + 16 x2 = 83', 'color': 'green', 'linestyle': 'dotted'},
    {'equation': lambda x: (234 - 10.3 * x) / 7.3, 'label': '10.3 x1 + 7.3 x2 = 234', 'color': 'yellow', 'linestyle': 'dashdot'}
]


fig, axes = plt.subplots(nrows=1, ncols=len(equations), sharey=True)


for i, eq in enumerate(equations):
    ax = axes[i]
    ax.plot(x, eq['equation'](x), color=eq['color'], linestyle=eq['linestyle'], label=eq['label'])
    ax.set_xlabel('x')
    ax.grid(True, linestyle=':', color='gray')
    ax.legend()


axes[0].set_ylabel('y')
plt.suptitle('Equation Lines')


plt.subplots_adjust(wspace=0.3)


plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-5, 5, 100)

# Define equations for the lines
equations = [
   {'equation': lambda x: (24 - 10.3 * x) / 8.4, 'label': '10.2 x1 – 3.2 x2 = 149', 'color': 'red', 'linestyle': 'dashed'},
    {'equation': lambda x: (203 - 3.2 * x) / 15.4, 'label': '– 5.8 x1 + 16 x2 = 83', 'color': 'green', 'linestyle': 'dotted'},
    {'equation': lambda x: (252 - 18.3 * x) / 3.2, 'label': '10.3 x1 + 7.3 x2 = 234', 'color': 'yellow', 'linestyle': 'dashdot'}
]

# Create a new figure with specified size and resolution
fig, ax = plt.subplots(figsize=(8, 8), dpi=100)

# Plot each line on the same figure
for eq in equations:
    ax.plot(x, eq['equation'](x), color=eq['color'], linestyle=eq['linestyle'], label=eq['label'])
    
    # Add labels for equations with suitable slopes
    slope = eq['equation'](1) - eq['equation'](0)  # Calculate slope using two points
    if abs(slope) > 1:
        x_label = 5
        y_label = eq['equation'](x_label)
        ax.text(x_label, y_label, eq['label'], fontsize=6, ha='center', va='center')

# Set x and y-axis labels
ax.set_xlabel('x1')
ax.set_ylabel('x2')

# Set the grid color and linestyle
ax.grid(True, linestyle=':', color='gray')

# Set the legend
ax.legend()

# Set the plot title
plt.title('Equation Lines')

# Display the plot
plt.show()
```
