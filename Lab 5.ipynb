##### Task 4 equation
\begin{cases}
        \begin{aligned}
            10.2 x_1 - 3.2 x_2 &= 149 \\
            -5.8 x_1 + 16 x_2 &= 83 \\
            10.3 x_1 + 7.3 x_2 &= 234 \\
        \end{aligned}
    \end{cases}

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4), sharey=True, sharex=True)


ax[0].plot(x, (149 - 10.2 * x) / -3.2, 'orangered', linestyle = '-.')
ax[1].plot(x, (83 + 5.8 * x) / 16,  'olivedrab', linestyle = ':')
ax[2].plot(x, (234 - 10.3 * x) / 7.3,  'mediumvioletred', linestyle = '--')

ax[0].set_title('10.2 x1  3.2 x2 = 149')
ax[1].set_title(' 5.8 x1 + 16 x2 = 83')
ax[2].set_title('10.3 x1 + 7.3 x2 = 234')

ax[0].grid(True, linestyle=':', color="darkgoldenrod")
ax[1].grid(True, linestyle=':', color="darkgoldenrod")
ax[2].grid(True, linestyle=':', color="darkgoldenrod")

ax[0].set_xlabel('x1')
ax[1].set_xlabel('x1')
ax[1].set_xlabel('x1')

ax[0].set_ylabel('x2')
ax[1].set_ylabel('x2')
ax[2].set_ylabel('x2')

plt.savefig('Task 1.png', dpi=300)
plt.savefig('Task 1.jpg', dpi=300)
plt.savefig('Task 1.svg', dpi=300)
```

```python
fig, ax = plt.subplots(figsize=(15,5), sharey=True, sharex=True)
x = np.linspace(5, 30, 100)

ax.plot(x, (149 - 10.2 * x) / -3.2, 'orangered', label= '10.2 x1  3.2 x2 = 149', linestyle = '-.')
ax.plot(x, (83 + 5.8 * x) / 16, 'olivedrab',label = ' 5.8 x1 + 16 x2 = 83', linestyle = ':')
ax.plot(x, (234 - 10.3 * x) / 7.3, 'mediumvioletred' ,label= '10.3 x1 + 7.3 x2 = 234', linestyle = '--')

equation1 = lambda x: (149 - 10.2 * x) / -3.2
equation2 = lambda x: (83 + 5.8 * x) / 16
equation3 = lambda x: (234 - 10.3 * x) / 7.3

ax.set_title('All Lines In One Figure')

ax.set_xlabel('x1')

ax.set_ylabel('x2')
ax.grid(True, linestyle=':', color="chocolate")
ax.legend()

x_label = 8
y_label = -25
label_text = '10.2 x1  3.2 x2 = 149'
ax.text(x_label, y_label, label_text, fontsize=12, ha='center', va='bottom', rotation=15)

x1_label = 8
y1_label = 8
label_text = ' 5.8 x1 + 16 x2 = 83'
ax.text(x1_label, y1_label, label_text, fontsize=12, ha='center', va='bottom')

x2_label = 8
y2_label = 17
label_text = '10.3 x1 + 7.3 x2 = 234'

ax.text(x2_label, y2_label, label_text, fontsize=12, ha='center', va='bottom', rotation= -7)

# INTERSECTION 1
eq1 = np.array([[10.2, -3.2], [-5.8, 16]])
v1 = np.array([149, 83])
intersection1 = np.linalg.solve(eq1, v1)
intersection_y = equation1(intersection1[0])
plt.scatter(intersection1[0], intersection_y, color='yellow', label='Intersection')


# INTERSECTION 2
eq2 = np.array([[-5.8, 16], [10.3, 7.3]])
v2 = np.array([83, 234])
intersection2 = np.linalg.solve(eq2, v2)
intersection_y = equation3(intersection2[0])
plt.scatter(intersection2[0], intersection_y, color='red', label='Intersection')

# INTERSECTION 3
eq3 = np.array([[10.2, -3.2], [10.3, 7.3]])
v3 = np.array([149, 234])
intersection3 = np.linalg.solve(eq3, v3)
intersection_y = equation3(intersection3[0])
plt.scatter(intersection3[0], intersection_y, color='green', label='Intersection')

# FILLING
x_fill = [intersection1[0], intersection2[0], intersection3[0]]
y_fill = [equation1(intersection1[0]), equation2(intersection2[0]), equation3(intersection3[0])]
ax.fill_between(x_fill, y_fill, color='green', alpha=0.3)

#  ARROWS

ax.annotate('the intersection between eq1 and eq3', xy=(17,8), xytext=(20,-50),
            arrowprops={'arrowstyle': '->'}, va='center', horizontalalignment='right', verticalalignment='top')

ax.annotate('the intersection between eq2 and eq3', xy=(15,10), xytext=(10,-50),
            arrowprops={'arrowstyle': '->'}, va='center', horizontalalignment='right', verticalalignment='top')

ax.annotate('the intersection between eq1 and eq2', xy=(18.3,10.1), xytext=(35,40),
            arrowprops={'arrowstyle': '->'}, va='center', horizontalalignment='right', verticalalignment='top')

#  SAVING AS AN IMAGE
plt.savefig('Task 2.png', dpi=300)
plt.savefig('Task 2.jpg', dpi=300)
plt.savefig('Task 2.svg', dpi=300)

plt.show()
```
